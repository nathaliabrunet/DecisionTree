```python
from __future__ import print_function
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn import tree
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import confusion_matrix , accuracy_score

pd.options.display.max_columns = 26
```

```python
#df = pd.read_excel('base.xlsx') #,error_bad_lines=False)
df=  pd.read_csv('df_tree.csv')
df.head(2)
```

```python
previsores = pd.get_dummies(df[['regiao','atividade','tipo']])
classe = df.loc[:,['classe']]
```

```python
classe.head()
```

```python
previsores['validade'] = df.validade
```

```python
print(previsores.columns)
```

```python
data_feature_names = list(previsores.columns)

class_names = list(df.classe.unique())
```

```python
previsores.head()
```

```python
#DECISIONTREECLASSIFIER
Tree = DecisionTreeClassifier(criterion='gini', splitter='best', max_depth=5)
Tree.fit(previsores,classe)
#predict = Tree.predict(previsores)
#tree.plot_tree(Tree.fit(previsores, classe))
#len(predict)
```

```python
from sklearn.externals.six import StringIO  
from IPython.display import Image  
from sklearn.tree import export_graphviz
import pydotplus
import graphviz
import collections
pd.options.mode.chained_assignment = None #None  # default='warn'

dot_data = StringIO()

export_graphviz(Tree, 
                out_file=dot_data,  
                feature_names = data_feature_names,
                class_names=class_names,
                filled=True, rounded=True,
                special_characters=True,
               impurity=False,
               label='root',
               proportion=False,
                precision = 3)
graph = pydotplus.graph_from_dot_data(dot_data.getvalue())  

#colors =  ('lightblue', 'lightyellow', 'forestgreen', 'lightred', 'white') ('salmon', 'aquamarine')
colors = ('darkviolet', 'cyan')
edges = collections.defaultdict(list)

for edge in graph.get_edge_list():
    edges[edge.get_source()].append(int(edge.get_destination()))

for edge in edges:
    edges[edge].sort()    
    for i in range(2):
        dest = graph.get_node(str(edges[edge][i]))[0]
        dest.set_fillcolor(colors[i])

#PNG
graph.write_png('tree.png')

#PDF
#graph_pdf = graphviz.Source(tree.export_graphviz(Tree, out_file=None)) 
#graph_pdf.render("tree")

Image(graph.create_png())
```
